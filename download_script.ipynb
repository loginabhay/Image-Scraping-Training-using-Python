```python
#ikea_scraper
import requests
import os
from bs4 import BeautifulSoup

pages = [
 'https://www.ikea.in/baby-children-products',
 'https://www.ikea.in/bathroom-sinksand-taps-cabinets',
 'https://www.ikea.in/beds',
 'https://www.ikea.in/chairs-stools-benches',
 'https://www.ikea.in/clothes-storage',
 'https://www.ikea.in/cookware',
 'https://www.ikea.in/decoration',
 'https://www.ikea.in/desks',
 'https://www.ikea.in/eating',
 'https://www.ikea.in/home-furnishing-accessories',
 'https://www.ikea.in/laundry-cleaning-products',
 'https://www.ikea.in/leisure-safety-products',
 'https://www.ikea.in/lighting',
 'https://www.ikea.in/mattresses',
 'https://www.ikea.in/mirrors',
 'https://www.ikea.in/outdoor-furniture',
 'https://www.ikea.in/small-storage-organisers',
 'https://www.ikea.in/sofas-armchairs',
 'https://www.ikea.in/storage-furniture',
 'https://www.ikea.in/tables',
 'https://www.ikea.in/tableware',
 'https://www.ikea.in/textilesand-curtains-rugs',
 'https://www.ikea.in/tools-fittings-curtain-rails',
 'https://www.ikea.in/wardrobes']
links = {}
for url in pages:
    try:
        response = requests.get(url)
        html = BeautifulSoup(response.content)
        content_gallery = html.find("div",{"class":"product_cat_gallery"})
        a_cg = content_gallery.findAll("a")
        links[url.split('/')[-1]] = [a.get("href") for a in a_cg]
    except Exception as e:
        print(e)
        print(url)
folder_path = './images'
os.mkdir(folder_path)
for key in links:
    os.mkdir(folder_path+"/"+key)
    count = 0 #for every category we are taking 1000 images or total available
    for item in links[key]:
        print(item)
        p = 1
        while count<1000:
            try:
                response = requests.get(item)
                if(response.status_code!=200):
                    break
                html = BeautifulSoup(response.content)
                images = html.findAll("div",{"class":"product-compact__image"})
                if(len(images)==0):
                    break
                for image in images:
                    for i in image.findAll("img"):
                        img = requests.get(i.get("src"))
                        print(img)
                        with open(folder_path + "/" +key+"/" + str(count) + ".jpg",'wb') as f:
                            f.write(img.content)
                        count+=1
                if(count<1000):
                    item = item+"/"+"page-"+str(p)
                    p+=1
            except Exception as e:
                print(e)
                break
```

```python
sum  = 0
for folder in os.listdir("./images"):
    if(len(os.listdir("./images/"+folder))==0):
        os.rmdir("./images/"+folder)
    else:
        sum += len(os.listdir("./images/"+folder))
print("total images: ",sum)
    
    
    
```
